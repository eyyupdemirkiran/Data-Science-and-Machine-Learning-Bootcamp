### ÖDEV 3: List Comprehension Applications

#### Görev 1: car_crashes verisindeki numeric değişkenlerin isimlerini büyük harfe çeviriniz ve başına NUM ekleyiniz.

```python
import seaborn as sns
df = sns.load_dataset("car_crashes")
df.columns
```

```python
df.columns = ["NUM" + col.upper() if df[col].dtype != "O" else col.upper() for col in df.columns]
df.columns
```

#### Görev 2: İsminde "no" BARINDIRMAYAN değişkenlerin isimlerininin SONUNA "FLAG" yazınız.

```python
import seaborn as sns
df = sns.load_dataset("car_crashes")
df.columns
```

```python
df.columns = [col.upper() + "_FLAG" if not "no" in col else col.upper() for col in df.columns]
df.columns
```

#### Görev 3: Aşağıda verilen değişken isimlerinden FARKLI olan değişkenlerin isimlerini seçerek yeni bir df oluşturunuz.

```python
import seaborn as sns
import numpy as np
import pandas as pd
df = sns.load_dataset("car_crashes")
df.columns
og_list = ["abbrev", "no_previous"]
```

```python
new_cols = [col for col in  df.columns if not col in og_list ]
new_cols
```

```python
np.array([str(row) for row in df[new_cols].values])
```

```python
new_df = pd.DataFrame(np.array([row for row in df[new_cols].values]), columns=new_cols)
new_df.head()
```
